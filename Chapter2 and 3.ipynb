### (1)创建一个数值范围0~1，间隔为0.01数组：

```python
import numpy as np
import os
```

```python
arr_t1 = np.arange(0,1,0.01)
print('The array in quired is:',arr_t1)
```

### (2)创建100个服从正态分布的随机数：

```python
arr_t2 = np.random.randn(100)
print('The array in quired is:',arr_t2)
```

### (3)对创建的两个数组进行四则运算

```python
arr_add = arr_t1 + arr_t2
arr_subtract = arr_t1 - arr_t2
arr_multiply = arr_t1 * arr_t2
arr_divide = arr_t1 / arr_t2
print(
    "add:",arr_add,"\n",
    "subtract:",arr_subtract,"\n",
    "multiply:",arr_multiply,"\n",
    "divide:",arr_divide,"\n",
)
```

## 2.分析1996~2015 人口数据特征分布、分散状况

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl

p = np.load('populations.npz')
times = np.flipud(p['data'][:-2,0]) 
total = np.flipud(p['data'][:-2,1]) 
total_male = np.flipud(p['data'][:-2,2]) 
total_female = np.flipud(p['data'][:-2,3]) 
total_town = np.flipud(p['data'][:-2,4]) 
total_village = np.flipud(p['data'][:-2,5]) 
names = p['feature_names']
label_gender = ['男','女']
label_location = ['城镇','乡村']


mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

### (2)每一幅子图绘制对应图形

#### 直方图

```python
p2 = plt.figure(figsize=(15,12),dpi = 400)

ax1 = p2.add_subplot(1,2,1)#创建子图：两行一列，选取第一张
plt.title('1996~2015 人口性别分布直方图')
plt.xlabel('年份')
plt.ylabel('人口（万人）')
plt.bar(times,total_male,color='blue',label='男',alpha=0.5)
plt.bar(times,total_female,color='pink',label='女',alpha=0.5)
plt.bar(times,total_male-total_female,color='red',label='男女人口差值',alpha=0.5)
plt.legend()


ax2 = p2.add_subplot(1,2,2)
plt.title('1996~2015 城乡人口分布直方图')
plt.xlabel('年份')
plt.ylabel('人口（万人）')
plt.bar(times,total_village,color='green',label = '乡村人口',alpha=0.5)
plt.bar(times,total_town,color='purple',label = '城镇人口',alpha=0.5)
plt.bar(times,total_town-total_village,color='yellow',label='城乡人口差值',alpha=0.5)
plt.legend()

plt.savefig('1996-2015 每5年人口性别及城乡人口数据直方图')
plt.show()
```

### （4）分析人口结构变化情况、变化速率增减状况

人口性别方面:
    从直方图中可见，近20年的记录中，男女人口皆保持着稳定的增长。男女间的差值2000年前逐渐拉大，之后近十年有所回落，保持稳定，2010年后差距又有所缩小。二十年间，男女数量总体差距虽然被稍微拉大，但2000年后差距一直保持收缩趋势，可推测出近年来女性的增长速率稍微高于男性。男女结构有慢慢平衡的倾向，
    从饼图中可见，以五年为分割，1996年的男女比例最接近平衡，但此后一段时间男性人口增长较快较多，即使近年女性人口占比保持回升态势，仍未能弥补之前的变化。总体上仍不如1996年。
    从箱线图中，男性人口上端相较下端集中，说明整体增长速率有所减缓。女性则上端相对下端分散，整体增长速率有所提升。
    
城乡人口方面:
    乡村人口与城镇人口在20年间呈此消彼长的状态。体现了人口结构开始伴随城市化的进程而改变。
